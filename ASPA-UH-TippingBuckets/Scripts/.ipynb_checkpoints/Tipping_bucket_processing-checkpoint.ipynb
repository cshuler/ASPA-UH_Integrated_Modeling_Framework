```python
# read in data and format the dates

files = os.listdir(raw_data_path) # make a list of all the files in the directory

for i in files:
    tempframe = pd.read_csv(os.path.join(raw_data_path, i), names =["Date", "Time", "Rain_in"])
    tempframe['DateTime'] = pd.to_datetime(tempframe['Date'] + ' ' + tempframe['Time'])
    del tempframe['Date'];    del tempframe['Time'] 
    tempframe['DateTime'] = pd.to_datetime(tempframe['DateTime'], errors='coerce')

    #  this applies the make-a-list-of-consecutive-dates function on the start and end dates from the data in the folder  
    start = tempframe['DateTime'][0]                                # first date in the list
    end =   tempframe['DateTime'].iloc[-1]                          # last date

    Bigdatelist = []
    for dt in datetime_range(start, end, {'minutes':1}):   # using above function make a list from start to end at whatever interval   
         Bigdatelist.append(dt)
    Big_date_frame = pd.DataFrame({'DateTime':Bigdatelist})
    Big_date_frame = Big_date_frame.merge(tempframe, on="DateTime", how = "outer")


    frame15min = Big_date_frame.set_index('DateTime').resample('15T').sum() 
    frame15min = frame15min.reset_index(drop=False)  


    fig = plt.figure(figsize=(12, 5))
    ax1 = plt.axes()
    ax1.set_title(i)
    plt.plot(frame15min['DateTime'], frame15min['Rain_in'], alpha = .7, color='b',  marker = '.', label = "15 minute flows [m3/s]")
```
